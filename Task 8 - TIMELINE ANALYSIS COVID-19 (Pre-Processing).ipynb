## DECLARING LIBRARY

```python
import pandas as pd
```

## LOADING DATASET

```python
covid= pd.read_csv('owid-covid-data.csv')
```

```python
covid
```

## CLEANING DATA

```python
# Only Extracting those columns which have only upto 20% value missing
cols = covid.columns[covid.isna().sum()/len(covid) <= 0.2]
covid_df = covid[cols]
covid_df
```

```python
covid_df.dropna(inplace=True)
```

## CREATING CLEAN DATASET

```python
covid_df.to_csv('Covid-19.csv', index=False)
```
