## **DECLARING LIBRARIES**

```python
import plotly.express as px
import plotly.graph_objects as go
from sklearn import datasets
import sklearn.metrics as sm
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## **LOADING DATASET**

```python
iris = datasets.load_iris()
```

```python
features = iris.data
features[:5]
```

```python
iris_class = iris.target_names
iris_class
```

```python
feature_names = iris.feature_names
feature_names
```

```python
labels = iris.target
labels
```

## *PLOTTING IRIS DATASET*

```python
df = px.data.iris()
fig = px.scatter(df, y="sepal_width", x="sepal_length", color="species",title='Iris Data')
fig.show('png')
```

## **CHECKING THE OUTLIERS AND SCALE OF DATASET**

```python
fig = go.Figure()
fig.add_trace(go.Box(y=df['sepal_length'],name=iris.feature_names[0]))
fig.add_trace(go.Box(y=df['sepal_width'],name=iris.feature_names[1]))
fig.add_trace(go.Box(y=df['petal_length'],name=iris.feature_names[2]))
fig.add_trace(go.Box(y=df['petal_width'],name=iris.feature_names[3]))
fig.show('png')
```

## **Pre-Processing**

**Split data into train and test set by 80-20 rule.**

```python
train_X,test_X,train_Y,test_Y = train_test_split(features,labels,test_size=0.20,random_state=0)
```

## **Training Model**

### **DECISION TREE CLASSIFIER**

```python
clf = tree.DecisionTreeClassifier().fit(train_X,train_Y)
```

## VISUALIZING DECISON TREE CLASSIFIER

**There are four factors,out of which one is used:**
- **Gini Index**
- **Information Gain**
- **Random Variation**
- **Chi Square**

**Factor used here is Gini Index and to calculate gini index the formulae is**
                **[Gini Index](http://www.learnbymarketing.com/481/decision-tree-flavors-gini-info-gain/)**

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(clf, 
               feature_names=feature_names,  
               class_names=iris_class,
               filled=True,
               rounded=True, 
               fontsize=14)
plt.show()
```

## **EVALUATING MODEL**

```python
pred_y=clf.predict(test_X)
```

```python
accuracy = sm.accuracy_score(test_Y,pred_y)
print('Classifier accuracy is {}%'.format(round(accuracy * 100),2))
```

```python
cf = sm.confusion_matrix(test_Y,pred_y)
print('Confusion Matrix:\n{}'.format(cf))
```
