## DECLARING LIBRARIES

```python
import pandas as pd
import numpy as np
```

## CREATING DATAFRAME

```python
terrorism_df = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='iso8859-1',low_memory=False)
```

```python
terrorism_df
```

```python
np.shape(terrorism_df)
```

## MANIPULATING DATA

```python
not_null_cols = terrorism_df.columns[terrorism_df.isna().sum()/len(terrorism_df) == 0]
```

```python
df = terrorism_df[not_null_cols]
```

```python
df
```

```python
df.info()
```

```python
np.shape(df)
```

```python
terrorism = df[(df['weaptype1_txt'] != 'Unknown')  &(df['region_txt'] != 'Unknown')
                & (df['country_txt'] != 'Unknown') &(df['dbsource'] != 'Unknown')
                &(df['property'] != 'Unknown') &(df['attacktype1_txt'] != 'Unknown')
                &(df['targtype1_txt'] != 'Unknown')]
```

```python
terrorism
```

```python
terrorism.drop(['eventid','country','region','weaptype1','targtype1',
                'attacktype1','INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY'],axis=1,inplace=True)
```

```python
terrorism.rename({'iyear':'year','imonth':'month',
                  'iday':'day','country_txt':'country',
                 'region_txt':'region', 'weaptype1_txt':'weapon type',
                 'targtype1_txt':'target', 'attacktype1_txt':'attack type',
                 'gname':'group name'},axis=1,inplace=True)
```

# SAVING DATAFRAME

```python
terrorism.to_csv(r'globalterrorism.csv', index = False)
```
